```python
data_path= r'C:/Users/zy/Desktop/source_img'
```

```python
def selname(file, target_file_name):

    for a in range(len(file[2])):
        datanames = file[2][a]
        # print("datanames",datanames)
        pattern = re.compile(target_file_name)
        match = pattern.search(datanames)
        if match != None:
            # print('return',file[2][a])
            return file[2][a]
```

```python


class Dataset():

    def __init__(self, path, counter_max=0, type='4D'):

        self.path = path
        self.class_folders = [folder for folder in os.listdir(
            self.path) if 'class' in folder]
        self.dataset = {'img': [], 'mask': []}
        self.patient_ids = []
        counter = 0

        for file in os.walk(path):

            if counter_max != 0 and counter > counter_max:
                break

            if file[0] == path:
                continue
            self.dataset['img'].append(os.path.join(
                file[0], selname(file, 'img')))  # mask
            self.dataset['mask'].append(
                os.path.join(file[0], selname(file, 'mask')))

            patient_id = os.path.basename(file[0])
            self.patient_ids.append(patient_id)
            print('/n')


data_dataset = Dataset(data_path)
```

```python

i = 0
feature_names = []
feature_values = []
```

```python

for x in range(len(data_dataset.patient_ids)):
    img = nib.load(data_dataset.dataset['img'][x]).get_fdata()
    msk = nib.load(data_dataset.dataset['mask'][x]).get_fdata()
    sitk_img = sitk.GetImageFromArray(np.array(img, dtype=np.int16))
    sitk_msk = sitk.GetImageFromArray(np.array(msk, dtype=np.int16))
    extractor = rm.featureextractor.RadiomicsFeatureExtractor()
    extractor.settings['label'] = 1
    features = extractor.execute(sitk_img, sitk_msk)  # from sitk image
    all_subjects = []
    for key, value in features.items():
        feature_names.append(key+'_data')  
        feature_values.append(value)
    df = pd.DataFrame([feature_values], columns=feature_names)


```
